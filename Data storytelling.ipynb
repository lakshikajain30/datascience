```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/Users/lakshikajain/Documents/GitHub/datascience/adult.csv')

# Ensure that the 'income' column is in a numeric format if it's not already
df['income'] = df['income'].replace({'<=50K': 0, '>50K': 1})

# Group by 'occupation',  'gender' to calculate the average income
income_diff = df.groupby(['occupation', 'gender'])['income'].mean().reset_index()

# Visualize the difference in income by gender within the same occupation
plt.figure(figsize=(12, 5))
sns.barplot(data=income_diff, x='occupation', y='income', hue='gender')
plt.title('Difference in Income by Gender in the Same Job')
plt.xticks(rotation=90)
plt.show()
```

```python
# Ensure that the 'income' column is in a numeric format if it's not already
df['income'] = df['income'].replace({'<=50K': 0, '>50K': 1})

# Group by 'occupation',  'gender' to calculate the average income
income_diff = df.groupby(['occupation', 'race'])['income'].mean().reset_index()

# Visualize the difference in income by gender within the same occupation
plt.figure(figsize=(12, 5))
sns.barplot(data=income_diff, x='occupation', y='income', hue='race')
plt.title('Difference in Income by Race in the Same Job')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualize the age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=False, bins=20)  # Histogram without KDE
plt.title('Distribution of Age by Number of Adults')
plt.xlabel('Age')
plt.ylabel('Number of Adults')
plt.show()
```

```python
# Group by 'workclass' and calculate the mean income
highest_salary_workclass = df.groupby('workclass')['income'].mean().reset_index()

# Identify the workclass with the highest average salary
highest_salary_workclass = highest_salary_workclass.sort_values(by='income', ascending=False)

print(highest_salary_workclass)

# Visualize the average salary by workclass
plt.figure(figsize=(10, 5))
sns.barplot(data=highest_salary_workclass, x='workclass', y='income')
plt.title('Average Salary by Workclass')
plt.show()

```

```python
income_by_marital_status = df.groupby('marital-status')['income'].mean().reset_index()

# Visualize the impact of marital status on income
plt.figure(figsize=(8, 4))
sns.barplot(data=income_by_marital_status, x='marital-status', y='income', palette='viridis')
plt.title('Impact of Marital Status on Income')
plt.xlabel('Marital Status')
plt.ylabel('Average Income (Proportion >50K)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by age to calculate the average income
income_by_age = df.groupby('age')['income'].mean().reset_index()
plt.figure(figsize=(10, 5))
sns.lineplot(data=income_by_age, x='age', y='income', marker='o', color='blue')
plt.title('Earnings Curve: Average Income by Age')
plt.xlabel('Age')
plt.ylabel('Average Income (Proportion >50K)')
plt.grid(True)
plt.show()
```
